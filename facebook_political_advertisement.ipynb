#Using for loops

```python
categorical_columns= []
for c in df.columns:
    if df[c].dtypes== object:
        categorical_columns.append(c)
categorical_columns
```

```python
numerical_columns= []
for n in df.columns:
    if n not in categorical_columns: 
        numerical_columns.append(n)
numerical_columns
```

```python
df[numerical_columns]
```

```python
df[numerical_columns].describe()
```

```python
plt.figure(figsize=(15,15))
for i, col in enumerate(numerical_columns):
    plt.subplot(4, 2, i+1)
    sns.boxplot(x=col, data=df)
plt.show()
```

```python
#using boxplot of seaborn
plt.figure(figsize=(15,15))
sns.boxplot(data=df, orient="h")
plt.show()
```

```python

# Assuming you already have a DataFrame named 'df' with numerical columns
plt.figure(figsize=(15, 15))

# Create a grid of subplots (2 rows and 1 column)
plt.subplot(2, 1, 1)
sns.boxplot(data=df, orient="h")
plt.title("Boxplots")

plt.subplot(2, 1, 2)
sns.histplot(data=df, kde=True)
plt.title("Histograms")

plt.tight_layout()  # Ensures proper spacing between subplots
plt.show()
```
